## Marketing data team analysis
### Based on the cost, analyse and give feedback on why some campaigns should be discontinued

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wrn
wrn.filterwarnings("ignore")
```

```python
df = pd.read_csv('Marketing Team Data - Sheet1.csv')
```

```python
df.head()
```

```python
df.info()
```

## Checking for missing values

```python
df.isnull().sum()
```

```python
df.dropna(axis=1, how='all',inplace=True)
df.head()
```

```python
df.columns = ["CAMPAIGN ID","CAMPAIGN NAME","AUDIENCE","AGE","GEOGRAPHY","REACH","IMPRESSIONS","FREQUENCY","CLICKS","UNIQUE CLICKS","UNIQUE LINK CLICKS","CLICK THROUGH RATE","UNIQUE CTR","AMOUNT SPENT ","COST PER CLICK","COST PER RESULT"]
df.head()
```

## Converting currency fields into float values

```python

currency_cols = ['AMOUNT SPENT ', 'COST PER CLICK', 'COST PER RESULT']

for col in currency_cols:
    df[col] = df[col].replace({'\$|,': ''}, regex=True).astype(float)
print(df.to_string())
```

## Checking for correlation

```python
df.corr(numeric_only=True)
```

```python
# ...existing code...

campaign_summary = df.groupby('CAMPAIGN NAME').agg({
    'AMOUNT SPENT ': 'sum',
    'CLICKS': 'sum',
    'IMPRESSIONS': 'sum',
    'REACH': 'sum',
    'CLICK THROUGH RATE': 'mean',
    'COST PER CLICK': 'mean',
    'COST PER RESULT': 'mean',
}).sort_values(by='AMOUNT SPENT ', ascending=False).reset_index()

# Display summary table
campaign_summary

# ...existing code...
```

## Visualization 
### Bar chart (The relationship between the campain name and the amount spent per campaign)

```python
plt.figure(figsize=(12, 6))
plt.bar(campaign_summary['CAMPAIGN NAME'], campaign_summary['AMOUNT SPENT '], color='orange')
plt.xlabel('CAMPAIGN NAME')
plt.ylabel('AMOUNT SPENT')
plt.title('Amount spent by each campaign')
plt.xticks(rotation=90)
plt.plot(campaign_summary['CAMPAIGN NAME'], campaign_summary['AMOUNT SPENT '], marker='o', color='blue', linewidth=2)
plt.show()
```

## Amount spent per campaign through reach

```python

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(campaign_summary['CAMPAIGN NAME'], campaign_summary['AMOUNT SPENT '], color='darkgreen', label='Amount Spent')
ax1.set_xlabel('CAMPAIGN NAME')
ax1.set_ylabel('AMOUNT SPENT', color='green')
ax1.tick_params(axis='y', labelcolor='green')
ax1.set_title('Amount spent per campaign through reach')
ax1.set_xticklabels(campaign_summary['CAMPAIGN NAME'], rotation=90)

ax2 = ax1.twinx()
ax2.plot(campaign_summary['CAMPAIGN NAME'], campaign_summary['REACH'], marker='o', color='red', label='REACH')
ax2.set_ylabel('REACH', color='red')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.show()
```

## Amount spent per campaign through Impressions

```python

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(campaign_summary['CAMPAIGN NAME'], campaign_summary['AMOUNT SPENT '], color='darkgreen', label='Amount Spent')
ax1.set_xlabel('CAMPAIGN NAME')
ax1.set_ylabel('AMOUNT SPENT', color='green')
ax1.tick_params(axis='y', labelcolor='green')
ax1.set_title('Amount spent per campaign through Impressions')
ax1.set_xticklabels(campaign_summary['CAMPAIGN NAME'], rotation=90)

ax2 = ax1.twinx()
ax2.plot(campaign_summary['CAMPAIGN NAME'], campaign_summary['IMPRESSIONS'], marker='o', color='red', label='IMPRESSIONS')
ax2.set_ylabel('IMPRESSIONS', color='red')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.show()
```

## Cost per click for every campaign 

```python
plt.figure(figsize=(15, 6))
plt.bar(campaign_summary['CAMPAIGN NAME'], campaign_summary['COST PER CLICK'], color='orange', label='COST PER CLICK (Bar)')

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(campaign_summary['CAMPAIGN NAME'], campaign_summary['COST PER RESULT'], color='orange', label='Cost Per Result (Bar)')
plt.plot(campaign_summary['CAMPAIGN NAME'], campaign_summary['COST PER RESULT'], marker='o', color='blue', linewidth=2, label='COST PER RESULT (Line)')
plt.xlabel('CAMPAIGN NAME')
plt.ylabel('COST PER RESULT')
plt.title('COST PER RESULT by each campaign')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
